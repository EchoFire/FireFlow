We now proceed to create lags for 7, 14, and 30 days. This means that:

We create a new column with the values of col shifted by lag days, within each tile_id group. That means:

It does not require to split the DataFrame.

It automatically handles the shifting only within each tile, avoiding data leakage across tiles.

The new column will have NaN for the first lag rows of each tile, which is expected.

Example: If one is using tavg and lag=7, this line will create tavg_lag7, which will contain the value of tavg from 7 days earlier (for the same tile).

```python
lags = [7, 14, 30] # days
for lag in lags:
    for col in ['max_T21', 'avg_NDVI', 'soil_moisture_am', 'soil_moisture_pm', 'tavg', 'tmax', 'prcp', 'wspd', 'wdir', 'pres']:
        mont_nf[f'{col}_lag{lag}'] = mont_nf.groupby('tile_id')[col].shift(lag)
```

Next step: We create rollings. This means:

We create a rolling average (or moving average) of column col over a defined window (e.g. 7, 14 days), within each tile (tile_id), and stores it in a new column named like 'col_roll7'.

üîç Explanation of each part: df.groupby('tile_id')[col] ‚Üí Groups the data by tile so we only compute rolling means within the same tile (no leakage between locations).

.transform(...) ‚Üí Applies the operation to each group and returns a series with the same length as the original DataFrame, aligning properly with it.

lambda x: x.shift(1) ‚Üí Shifts the values one day, takes the previous day, so the value of the day itself is not taken into the average.


x.rolling(window).mean() ‚Üí This is an anonymous function (a quick function with no name) applied to each tile group. It does:

x.rolling(window) ‚Üí Defines a rolling window of size window over the series x.

.mean() ‚Üí Computes the mean over each window.

Example: If you have:

col = "tavg"

window = 7

It will create tavg_roll7, where each value is the average of the temperature of the 7 previous days on that tile.

lambda = quick inline function.

```python
windows = [7, 14]
for window in windows:
    for col in ['max_T21', 'avg_NDVI', 'soil_moisture_am', 'soil_moisture_pm', 'tavg', 'tmax','prcp', 'wspd', 'wdir', 'pres']:
        mont_nf[f'{col}_roll{window}'] = mont_nf.groupby('tile_id')[col].transform(lambda x: x.shift(1).rolling(window).mean())
```

# Creating the data for the Park
We aggregate the data of the tiles by date. The criteria is different depending of the variable.

```python
montes2 = mont_nf.copy()
```

We drop the columns 'tile_id' and 'wdir', wind direction, because it does not make sense to keep them when we aggregate the tiles by date.

```python
montes2 = montes2.drop(['tile_id'], axis=1)
```

```python
montes2 = montes2.drop(columns=[col for col in montes2.columns if 'wdir' in col])
```

```python
montes2.columns
```

In order to perform the aggregation in an organized form, we create a dictionary that will give the criteria to follow in each variable when doing the aggregation. For some variables we will consider the maximum value, this is the value that we will have for the park on a given day is the maximum over the tiles on that day. For others we consider the minimum and for others the average. The choice was made logically but also taking into account what would be the best for predicting appearances of fires.

```python
# 1. Determine the criteria for the aggregation of each feature
cols_max = [col for col in montes2.columns if 'max_T21' in col or 'avg_NDVI' in col or 'tmax' in col]
cols_min = [col for col in montes2.columns if 'soil_moisture' in col]
cols_other = [col for col in montes2.columns
              if col not in ['date'] + cols_max + cols_min]
cols_max, cols_min, cols_other
```

```python
# 2. Build the aggregation dictionary
agg_dict2 = {}
for col in cols_max:
    agg_dict2[col] = 'max'
for col in cols_min:
    agg_dict2[col] = 'min'
for col in cols_other:
    agg_dict2[col] = 'mean'
```

```python
agg_dict2
```

```python
montes=montes2.copy()
```

```python
montes = montes.groupby('date').agg(agg_dict2).reset_index()
```

```python
montes.info()
```

```python
montes['date'].min(), montes['date'].max()
```

```python
montes.head(40)
```

Now we fill missing values. There are only the created features missing. After verfiying that none of the maxT21 related first valid values was different from 0, we fill the missing values by the next valid value.

```python
# Fill lag and roll columns using backward fill (next valid value)
lag_roll_cols = [col for col in montes.columns if 'lag' in col or 'roll' in col]
montes[lag_roll_cols] = montes[lag_roll_cols].bfill()
```

## Montesinho by week

We procced now to aggregate the data by week in order to have a data set with weekly values. This is to increase the proportion of fires in the data set.

```python
montes_20_23_week = montes_2020_2023.copy()
```

```python
# Step 1: Create 'year_week' column
montes_20_23_week['year_week'] = montes_20_23_week['date'].dt.strftime('%Y-%W')
# %W is week number (starting Monday)
```

```python
montes_20_23_week['year_week'].head(40)
```

```python
montes_20_23_week['month'] = montes_20_23_week['date'].dt.to_period('M').dt.to_timestamp()
```

```python
# Step 2: Drop 'date'
montes_20_23_week = montes_20_23_week.drop(columns=['date'])
```

```python
montes_20_23_week.info()
```

Now we create a dictionary to make the aggregation by week

```python
# 1. Determine the criteria for the aggregation of each feature
cols_max = [col for col in montes_20_23_week.columns if 'max_T21' in col or 'avg_NDVI' in col or 'tmax' in col or 'month' in col]
cols_min = [col for col in montes_20_23_week.columns if 'soil_moisture' in col]
cols_other = [col for col in montes_20_23_week.columns
              if col not in ['year_week'] + cols_max + cols_min]
cols_max, cols_min, cols_other
```

```python
# 2. Build the aggregation dictionary per week
agg_dict_week = {}
for col in cols_max:
    agg_dict_week[col] = 'max'
for col in cols_min:
    agg_dict_week[col] = 'min'
for col in cols_other:
    agg_dict_week[col] = 'mean'
```

```python
agg_dict_week
```

```python
# Step 3 & 4: Group by and aggregate
montes_week_20_23 = montes_20_23_week.groupby('year_week').agg(agg_dict_week).reset_index()
```
