#Merging data from different sources into a single data set
Here we merges the csv data sets corresponding to fire, ndvi, soil moisture and weather conditions for Portugal with the tessellation. Creates a big data set for the period of time 01.01.2020 until 31.03.2025, and also single data sets with the data of the years 2020, 2021, 2022, 2023, and 2024.

```python
wp['time']= pd.to_datetime(wp['time'])
```

```python
wp.info()
```

```python
# I make copies of the data frames before deleting and merging
fire_P = fp.copy()
ndvi_P = ndvip.copy()
sm_P = smp.copy()
weather_P = wp.copy()
```

Now, I start deleting the .geo columns in the fire, soil moisture and ndvi data frames before merging.

```python
fp = fp.drop(['.geo'], axis=1)
ndvip = ndvip.drop(['.geo'], axis=1)
smp = smp.drop(['.geo'], axis=1)
```

We drop the columns that are going to be repeated.

```python
ndvip = ndvip.drop(['centroid_alt', 'centroid_lat', 'centroid_lon'], axis=1)
smp = smp.drop(['centroid_alt', 'centroid_lat', 'centroid_lon'], axis=1)
```

```python
ndvip.info()
```

```python
first_date_wc = weather_P['date'].min()
last_date_wc = weather_P['date'].max()
first_date_wc, last_date_wc
```

```python
df_all = fp.merge(ndvip, on=['tile_id', 'date'], how='left', sort=False) \
            .merge(smp,  on=['tile_id', 'date'], how='left', sort=False) \
            .merge(wp,  on=['tile_id', 'date'], how='left', sort=False)
```

```python
df_all.info()
```

```python
df_all.sample(30)
```

```python
df_all = df_all.drop(['system:index_x', 'system:index_y', 'system:index'], axis=1)
```

```python
df_all['max_T21'].describe()
```

```python
mask_1 = df_all['max_T21']>0
min_val = df_all.loc[mask_1, 'max_T21'].min()
max_val = df_all.loc[mask_1, 'max_T21'].max()
min_val, max_val
```
